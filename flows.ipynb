```python

from google.cloud import dialogflowcx_v3


def list_flows():
    # Create a client
    client = dialogflowcx_v3.FlowsClient()

    # Initialize request argument(s)
    request = dialogflowcx_v3.ListFlowsRequest(
        parent="projects/bot1-404120/locations/global/agents/79c5a35e-32a9-403b-ad36-d24f4f7d5d67",
    )

    # Make the request
    page_result = client.list_flows(request=request)

    # Handle the response
    for response in page_result:
        print(response)
```

```python
%%capture cap --no-stderr



print(list_flows())
with open('output.txt', 'w') as f:
    f.write(cap.stdout)
```
